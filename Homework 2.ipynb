In this program we are asking for the age of 3 students in order to obtain the sample mean, we can calcultate it just adding the three numbers and dividing by 3, in this case is 3 because we just are asking for three ages


```python
x=f"Welcome please type the ages of the students max value 90"
print(x)           #another way to print messages in this case it was storaged in a variable
n1=int(input("Writte the age of the fisrt student: ")) #here we ask for the ages
n2=int(input("Writte the age of the second student: ")) 
n3=int(input("Writte the age of the tird student: ")) 

s=(n1+n2+n3)/3  #and for the sample mean we just add the three ages and divide by the number of age in this case 3
print(f"The sample mean is {s}")  
```
