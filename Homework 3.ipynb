In this program we ask for the quantity of numbers that the list will give, after that we will put the names(the number of the names depend of the numbers of the list) then it will show us the list and it will ask us to choose our favorite name, if the number If the number entered is greater than the one on the list, it will send us an error.

```python
v= int(input("How many names do you want to introduce: ")) #here we ask for the quantity of numbers in the list
names = []
total = 1
for i in range (0,v):
    x=input("Introduce names: ") #here we put the names
    names.append(x)
    
print("The names:") #this just is for show the list of the names

for name in names:
    print(name)
f=int(input("Whats your favorite?: ")) #here we choose a name 
if f>v or f<0:
    print("Error, Please retype the number") #if the number is exceed the number of the number of numbers will give error

else:
    print(names[f])
```
