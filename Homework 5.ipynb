In this program we ask for the trimmed mean the fist step is to ask for the total data, 
after that we need to accomodate the data, ask for the percentage  and finally apply the formulas


```python
    x=int(input("How many data you will use?:  "))#here we ask for the total data

    my_list=[]
    for i in range(0,x):
        m=float(input(f"Introduce the data {i+1}: "))
        my_list.append(m)

    my_list = sorted(my_list) #with this command we accomodate the data
    print ("The ordered list is:", my_list)
    p=int(input("The percentage is: "))     #here we ask for the percentage
    pf=(p/100)*x                  
    u = my_list[int(pf):-int(pf)]      
    print ("These numbers trimmed are:",u)
    print ("The trimmed mean is:",sum(u)/len(u))
                    #Memebers of the team- Vargas Perez Josue Miguel
```
