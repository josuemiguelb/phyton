```python

print("---------MENU--------")#The menu of the program
print("1.Sample Mean")
print("2.Moda")
print("3.Standard deviation")
numero =int(input("Please select the option: "))
if numero == 1:
    val =int(input("Introduce the number of data you will use: "))
    data=[]
    total=0  
    for i in range (0,val):
        x=int(input(f"digitize the  data {i+1}: "))
        data.append(x)
        total+=x        #here we add the  data (x)
        sample_mean=total/val       #with this formula we obtain the sample mean
       
    print("the sample mean is: ",sample_mean)
    
    
elif numero == 2:
    print("-----------MODA-----------------")
    val= int(input("Introduce the number of data you will use: "))
    data=[]
    total=0  
    for i in range (0,val):
        x= int(input(f"digitize the  data {i+1}: "))
        data.append(x)
    
        rep = 0                                                                         
    for i in data:                                                                              
        appears = data.count(i)                                                             
        if appears > rep:                                                       
            rep = appears                                                       
                                                                                         
    modas = []                                                                               
    for i in data:                                                                              
        appears = data.count(i)                                                             
        if appears == rep and i not in modas:                                   
            modas.append(i)  
        
                                                                                         
    print ("Moda is: ", modas)

elif numero ==3:
    print("--------------STANDARD DEVIATION----------")
    val =int(input("Introduce the number of data you will use: "))
    data=[]
    total=0 
    
    for i in range (0,val):
        x=int(input(f"digitize the  data {i+1}: "))
        data.append(x)
        total+=x
        sample_mean=total/val
        squares = [] #list of the sum of the squares of the differences beetween the data and the mean
        for dato in data:
          r = (dato - sample_mean)**2
          squares.append(r)
        
        desviation = (sum(squares)/(val - 1))**0.5 #this is the formula of standard deviation(mean)
    
    print ("The Standard deviation is: ", desviation)
    
else:
    print("ERROR,RETYPE A NUMBER")
    #names of the members: Vargas Perez Josue Miguel, Aquino del Angel Oswaldo

```
